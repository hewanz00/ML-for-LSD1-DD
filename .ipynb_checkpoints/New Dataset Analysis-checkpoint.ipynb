```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# if using a Jupyter notebook, includue:
%matplotlib inline
```

```python
#cleansed new data added
```

```python
data = pd.read_csv("./datasets/new dataset.csv")
data
```

New data statistics collection

```python
pchembl_values = data["pchembl values"].tolist()
pchembl_values
```

mean

```python
np.mean([pchembl_values])
```

std

```python
np.std([pchembl_values])
```

```python
print(np.var([pchembl_values]))
```

```python
print(np.median([pchembl_values]))
```

New dataset histogram

```python
pchembl = data["pchembl values"]
pchembl
```

```python
pchembl.hist(grid=True, bins=12, rwidth=0.9,
                   color='#FF7F24')
plt.title('pCHEMBL distribution')
plt.xlabel('pCHEMBL')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt. grid(False)
```

Original dataset histogram

```python
data2 = pd.read_csv("datasets/3_512_y_main.csv")
data2
```

```python
data2.hist(grid=True, bins=12, rwidth=0.9,
                   color='#3D59AB')
plt.title('pCHEMBL distribution')
plt.xlabel('pCHEMBL')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt. grid(False)
```

Both dataset histograms together

```python
data2.hist(grid=True, bins=12, rwidth=0.8,
                   color='#3D59AB', label='original pCHEMBL')
pchembl.hist(grid=True, bins=12, rwidth=0.7,
                   color='#FF7F24', label='new pCHEMBL')
plt.title('pCHEMBL distribution')
plt.xlabel('pCHEMBL')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.5)
plt. grid(False)
plt.savefig("figures/histogram.png")
```
